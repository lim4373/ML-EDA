```python
# 함수 Null 처리 함수
from sklearn.preprocessing import LabelEncoder


def fillna(df):
  df['Age'].fillna(df['Age'].mean(),inplace=True)
  df['Cabin'].fillna('N',inplace=True)
  df['Embarked'].fillna('N',inplace=True)
  df['Fare'].fillna(0,inplace=True)
  return df

def drop_features(df):
  df.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
  return df

def format_features(df):
  df['Cabin']=df['Cabin'].str[:1]
  features = ['Cabin','Sex','Embarked']
  for feature in features:
    le = LabelEncoder()
    le = le.fit(df[feature])
    df[feature]=le.transform(df[feature])
  return df

def transform_features(df):
  df = fillna(df)
  df = drop_features(df)
  df = format_features(df)

  return df

```

```python
transform_features(titanic_df)
```

```python
X_titanic_df=titanic_df.drop('Survived',axis=1)
y_titanic_df=titanic_df['Survived']

X_titanic_df = transform_features(X_titanic_df)
```

#Gridsearchcv

```python
from sklearn.model_selection import GridSearchCV

xgb_clf = xgb.XGBClassifier()
params = {
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [100, 300, 500],
    'max_depth': [3, 5, 7],
}

grid_cv = GridSearchCV(xgb_clf, param_grid=params, cv=3, scoring='accuracy')
grid_cv.fit(X_train, y_train)

print('최적 하이퍼 파라미터:', grid_cv.best_params_)
print('최고 예측 정확도: {0:.4f}'.format(grid_cv.best_score_))

# 테스트 데이터로 성능 평가
xgb_preds = grid_cv.best_estimator_.predict(X_test)
print('테스트 세트 정확도: {0:.4f}'.format(accuracy_score(y_test, xgb_preds)))
```
